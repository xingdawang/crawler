```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

book_tag = '计算机'
book_number = 0
each_page_books = 20
headers = {'User-Agent': 'Mozilla/5.0 (Linux; Android 5.1.1; SM-G928X Build/LMY47X) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.83 Mobile Safari/537.36'}


# extract features
book_columns = ['book_link', 'book_profile_image', 'book_title', 'book_pub', 'book_rating', 'book_rating_number', 'book_intro']

# setup a dictionary structure to store information.
book_structure_dict = {}
for column in book_columns:
    book_structure_dict[column] = []

# douban has a bug, show more books than actual it has.
while(True):
    url = 'https://book.douban.com/tag/' + book_tag + '?start=' + str(book_number) + '&type=T'
    
    page = requests.get(url, headers = headers)
    soup = BeautifulSoup(page.text, 'html.parser')
    
    book_list = soup.find('div', {'id': 'subject_list'})

    books = book_list.find_all('li')

    # if this page has no book, it means you read all book, douban shows more books than actual it has.
    if not books:
        break

    # get all books in current page
    for book in books:

        # book link and book profile image
        book_info = book.contents[1]

        # book link
        book_link = book_info.contents[1].get('href')
        book_structure_dict['book_link'].append(book_link)

        # book profile image
        book_profile = book_info.contents[1]
        book_profile_image = book_profile.find('img').get('src')
        book_structure_dict['book_profile_image'].append(book_profile_image)

        # book details
        book_details = book.contents[3]

        # book title
        book_title = book_details.find('a').get('title')
        book_structure_dict['book_title'].append(book_title)

        # book publication info
        book_pub = book_details.find('div', {'class': 'pub'}).getText().strip()
        book_structure_dict['book_pub'].append(book_pub)

        # book rating
        book_rating = book_details.find('span', {'class': 'rating_nums'})
        if book_rating != None:
            book_rating_content = book_rating.getText()
        else:
            book_rating_content = None
        book_structure_dict['book_rating'].append(book_rating_content)

        # rating number
        book_rating_number = book_details.find('span', {'class': 'pl'}).getText().strip()
        book_structure_dict['book_rating_number'].append(book_rating_number)

        # book intro
        book_intro = book_details.find('p')
        if book_intro != None:
            book_intro_content = book_intro.getText()
        else:
            book_intro_content = None
        book_structure_dict['book_intro'].append(book_intro_content)

    print('Read No.%d page.' % int(book_number / 20 + 1))
    book_number += each_page_books

result = pd.DataFrame.from_dict(book_structure_dict)
result_name = book_tag + '.csv'
result.to_csv(result_name, index=False)
```
